```python
from sklearn.model_selection import train_test_split
```

```python
#Predicting High column
X=df.drop(["CLOSE","DATE","TIME"],axis=1)
y=df["CLOSE"]
```

```python
# Trainset>> X_train,y_train # Test Set >>X_test,y_test 

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mae = mean_absolute_error(y_test, y_pred)
```

```python

mse = mean_squared_error(y_test, y_pred)
```

```python

print('mae : ',mae)
print('mse : ',mse)
```

```python
data_point = [[13528.1340, 13528.1340, 13481.0880, 3536]]
#The real value of close that we will predict = 13481.0880
# Convert data points to numeric format and reshape them
new_data = np.array(data_point)

# Make predictions
predictions = lr.predict(new_data)

# Print predictions
print(predictions)

```

```python
from sklearn.ensemble import RandomForestRegressor

# Create a Random Forest Regression model
rf_model = RandomForestRegressor(n_estimators=150, random_state=80)  # You can adjust the number of estimators (trees)

# Fit the model to the training data
rf_model.fit(X_train, y_train)

# Make predictions
rf_predictions = rf_model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, rf_predictions)
mse = mean_squared_error(y_test, rf_predictions)

print('mae : ',mae)
print('mse : ',mse)
```

```python
predictions = rf_model.predict(new_data)
print("Prediction of CLOSE ",predictions)
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create an SVR model
svr_model = SVR(kernel='rbf')  # You can specify other kernels like 'linear', 'poly', etc.

# Fit the model to the scaled training data
svr_model.fit(X_train_scaled, y_train)

# Make predictions on the scaled testing data
svr_predictions = svr_model.predict(X_test_scaled)
```

```python
mae = mean_absolute_error(y_test, svr_predictions)
mse = mean_squared_error(y_test, svr_predictions)

print('mae : ',mae)
print('mse : ',mse)
```

```python
predictions = svr_model.predict(new_data)
print(predictions)
```

```python
from sklearn.tree import DecisionTreeRegressor

# Create a Decision Tree Regression model
dt_model = DecisionTreeRegressor(random_state=42)  # You can specify hyperparameters like max_depth, min_samples_split, etc.

# Fit the model to the training data
dt_model.fit(X_train, y_train)

# Make predictions
dt_predictions = dt_model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, dt_predictions)
mse = mean_squared_error(y_test, dt_predictions)

print('mae : ',mae)
print('mse : ',mse)
```

```python
predictions = dt_model.predict(new_data)
print(predictions)
```
